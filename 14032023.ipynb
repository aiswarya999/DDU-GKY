# Query

### Boarded from southampton 

```python
boarded_southampton = df.query("Boarded == 'Southampton'")
```

```python
boarded_southampton
```

### Destination at newyork

```python
destination_newyork = df.query("Destination == 'New York City'")
```

```python
destination_newyork
```

#### Boarded from southampton and destinated at Newyorkcity

```python
destination_newyork_boarded_southampton = boarded_southampton.query("Destination == 'New York City'") 
```

```python
destination_newyork_boarded_southampton
```

#### Females boarded from southampton and destinated at newyorkcity

```python
destination_newyork_boarded_southampton_female = destination_newyork_boarded_southampton.query("Sex == 'female'")
```

```python
destination_newyork_boarded_southampton_female
```
