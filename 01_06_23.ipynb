```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
data=pd.read_csv('./titanic (1).csv')
data.head()
```

## Groupby

### Genderwise survival

```python
sex_survived_sum=data.groupby(['Sex'])['Survived'].sum()
sex_survived_sum
```

```python
Embarked_survived_sum=data.groupby(['Embarked'])['Survived'].sum()
Embarked_survived_sum
```

```python
Embarked_age_sum=data.groupby(['Embarked'])['Age'].sum()
Embarked_age_sum
```

```python
sexsurvived_Age_average=data.groupby(['Sex','Survived'])['Age'].mean()*100
sexsurvived_Age_average
```

```python
pclass_survived_sum=data.groupby(['Pclass'])['Survived'].sum()
pclass_survived_sum
```
