### Feature engineering the medal column

```python
data.Medal.unique()
```

### Lets create seoarate columns using if_else statement for number of medals 

```python
data['Gold']=[1 if medal == 'Gold' else 0 for medal in data['Medal']]
```

```python
data['Silver']=[1 if medal == 'Silver' else 0 for medal in data['Medal']]
data['Bronze']=[1 if medal == 'Bronze' else 0 for medal in data['Medal']]
```

```python
data.head()
```

```python
data.drop(columns=['ID','Games'], inplace=True)
```

```python
data.head()
```

### Which country has won the most number of medals

```python
country_wise_medal = data.groupby(['Team'])['Gold','Silver','Bronze'].sum()
country_wise_medal
```

### Lets add a new column to the groupby object

```python
country_wise_medal['Total'] =country_wise_medal['Gold']+country_wise_medal['Silver']+country_wise_medal['Bronze']
country_wise_medal
```

```python
country_wise_medal.sort_values(by = 'Total', ascending=False)
```

```python
country_wise_medal.nlargest(1,'Total')
```

```python
country_wise_medal.nsmallest(1,'Total')
```

### Country with no gold medal at all

```python
No_gold_medals=country_wise_medal[country_wise_medal['Gold']==0]
No_gold_medals
```

```python
len(No_gold_medals)
```

```python
List_No_gold_medals=list(No_gold_medals.index)
List_No_gold_medals
```

```python
print("Number of countries with no Gold medals  :", len(List_No_gold_medals))
```
