### Lets add new columns from the date column

```python
data['Year'] = data.Date.dt.year
data['Month'] = data.Date.dt.month
data['Week_day'] = data.Date.dt.weekday
data
```

```python
data['Month'] =data.Month.replace({1:'Janurary',
                                   2:'February',
                                   3:'March',
                                   4:'April',
                                   5:'May',
                                   6:'June',
                                   7:'July',
                                   8:'August',
                                   9:'Sepetember',
                                   10:'October',
                                   11:'November',
                                   12:'December'})
data['Week_day']=data.Week_day.replace({0:'Sunday',1:'Monday',2:'Tuesday',3:'Wednesday',4:'Thurdsay',5:'Friday',6:'Saturday'})
data
```

### Temperature conversion(-2&#176;F-32)* 5/9

```python
data['Celsius'] = [(Temp-32)*(5/9) for Temp in data.Temperature]
```

### Lets create a new column called Heat

```python
data['Heat'] = pd.cut(x=data.Celsius , bins=[-20,10,35], labels=['cold','warm'])
data
```

```python
sns.barplot(x='Heat', y='Weekly_Sales', hue='Holiday_Flag',data=data)
plt.show()
```
