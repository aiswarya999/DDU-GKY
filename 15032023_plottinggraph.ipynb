# Univariate Analysis

## KDE plot

```python
plt.figure(figsize=(4,3))
sns.kdeplot(data=data.PassengerId)
plt.show()
```

```python
plt.figure(figsize=(10,4))
sns.kdeplot(data=data.Age_wiki)
plt.show()
```

## HISTPLOT

```python
sns.histplot(data.Fare)
plt.show()
```

## BOX PLOT

```python
sns.boxplot('Age', data=data)
plt.show()
```

# Multivariate Analysis

## Line Plot

```python
sns.lineplot('Age','Fare', data=data)
plt.title("Age vs Fare")
plt.show()
```

```python
sns.lineplot('Pclass','Age', data=data)
plt.title("Pclass vs Age")
plt.show()
```

```python
sns.lineplot('Pclass','Fare', data=data)
plt.title("Pclass vs Fare")
plt.show()
```

# PIE CHART

```python
data.columns
```

```python
pclass_survived = data.groupby(['Pclass'])['Survived'].sum()
pclass_survived
```

```python
sns.set_style('ticks')
pclass_survived.plot.pie()
plt.legend()
plt.show()
```

```python
pclass_sex_survived = data.groupby(['Pclass','Sex'])['Survived'].sum()
```

```python
pclass_sex_survived.reset_index()
```

```python
pclass_sex_survived.plot.pie(autopct = '%1.2f%%')
plt.legend(bbox_to_anchor=(1.5,1), loc='upper left', borderaxespad=0)
plt.show()
```

```python
pclass_sex_survived.plot.pie(autopct = '%1.2f%%')
plt.legend(bbox_to_anchor=(1.5,1), loc='lower left', borderaxespad=0)
plt.show()
```

```python
pclass_sex_survived.plot.pie(autopct = '%1.2f%%')
plt.legend(bbox_to_anchor=(1.5,1), borderaxespad=0)
plt.show()
```

```python
pclass_sex_survived.plot.pie(autopct = '%1.2f%%')
plt.legend(bbox_to_anchor=(1.5,1))
plt.show()
```

```python
pclass_sex_survived.plot.pie(autopct = '%1.4f%%')
plt.legend(bbox_to_anchor=(1.5,1))
plt.show()
```

## Bar Chart

```python
plt.figure(figsize=(4,3))
sns.countplot('Sex', hue = 'Survived', data=data)
plt.show()
```

## Count plot

```python
plt.figure(figsize=(4,3))
sns.countplot('Survived', data=data)
plt.show()
```

```python
plt.figure(figsize=(4,3))
sns.set_style('darkgrid')
sns.countplot('Pclass',hue='Survived', data=data)
plt.title('Pclass: Survived vs Dead')
plt.show()
```

## Factor Plot

```python
sns.set_style('ticks')
sns.factorplot('Pclass','Survived', hue='Sex', data=data)
plt.show()
```
